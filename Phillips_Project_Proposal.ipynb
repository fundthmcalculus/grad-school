````python
### BEGIN SOLUTION
data = pd.read_csv('turbine-data.csv')
# Remove columns that aren't relevant (eg metadata)
# ControlBoxTemperature has 2 values, WTG has 1.
data = data.drop(columns=['ControlBoxTemperature','WTG'])

# Drop all non-numeric columns
data = data.select_dtypes(include=[np.number]).copy()

# Dropping NA here to ensure enough rows.
data = data.dropna().copy()

### END SOLUTION
print("Number of rows:", data.shape[0])
print("Number of columns:", data.shape[1])
````

##Step 3 — 🧹 Label Cleanup [5 points]

Your regression label should be a continuous and clean variable with a reasonably uniform distribution.

Before proceeding, make sure your label column contains valid and meaningful values.

1. **Select the original label column name** and store it in a variable named `label_name`.
2. **Remove records** where the label is missing (NaN) or invalid.
3. **Filter out extreme or rare label values** — for example, if your label histogram shows very few samples at the far ends (too low or too high), remove those records.
4. Store the final cleaned continuous label in `data['Regression_Label']`.

Cleaning your label helps ensure the project is suitable for regression and makes later classification binning more meaningful.


````python
### BEGIN SOLUTION

# Example:
# label_name = 'price'
# data = data[data[label_name].notna()].copy()
# data = data[(data[label_name] >= lower_bound) & (data[label_name] <= upper_bound)].copy()

label_name = 'ActivePower'
data['Regression_Label'] = data[label_name]
# Because the histogram is imbalanced with 0s, remove those
lower_limit = 100
upper_limit = 1600 # There's a lot at ~1750
data = data[(data['Regression_Label'] > lower_limit) & (data['Regression_Label'] < upper_limit)].copy()

### END SOLUTION
````

````python
#If you have completed the previous step correctly, running the following command should generate a histogram of your labels.
data['Regression_Label'].hist()
````

##🧩 Step 4 — Feature Selection [5 points]

In this step, you will **analyze and select** the most relevant features for your regression model.

1. **Write and run your own code** to examine the linear correlations among features — for example, by calculating and visualizing a correlation matrix (using `corr()` or a heatmap).
2. Based on your analysis, **identify and remove highly correlated features** to reduce redundancy.
3. Store the removed feature names in a Python list called `removed_features`.
4. Store your final selected features in a DataFrame named `X`.

⚠️ Important:
- Do **not** include `Regression_Label` or `Classification_Label` in `X`.
- `X` should contain the final cleaned feature set you want to use for prediction.


````python
### BEGIN SOLUTION

# Example:
# removed_features = ['feature_a', 'feature_b']
# X = data[['feature1', 'feature2', 'feature3']]

# Get all numeric features except the labels
feature_cols = [col for col in data.columns if col not in ['Regression_Label', 'Classification_Label', label_name]]
features_df = data[feature_cols]

# Calculate correlation matrix
corr_matrix = features_df.corr()

# Visualize correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# Identify highly correlated features (threshold of 0.85)
removed_features = []
correlation_threshold = 0.85

for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > correlation_threshold:
            colname = corr_matrix.columns[j]
            if colname not in removed_features:
                removed_features.append(colname)

print(f"Removed features due to high correlation: {removed_features}")

# Create final feature set
final_features = [col for col in feature_cols if col not in removed_features]
X = data[final_features]

### END SOLUTION
````

````python
print("Number of Features selected: ", X.shape[1])
````

##🧮 Step 5 — Modify the Regression Problem into a Classification Problem [5 points]

The dataset you selected is originally a **regression dataset**, meaning its label represents a **continuous numeric value** (for example, car price, temperature, or stock price).

In this step, you’ll **transform the continuous label into discrete categories**, effectively converting the problem from **regression** to **classification**.

To do this, you will **bin** the numeric label values into groups (categories).
For example, if your label ranges from 1 to 100, you could group it as follows:
*   Values between 1–20 → Category 1 (bin size: 20)
*   Values between 20–30 → Category 2 (bin size: 10)
*   Values between 30–60 → Category 3 (bin size: 30)
*   Values between 60–100 → Category 4 (bin size: 40)

⚠️ **Important**:
*   Bins do not have to be equal in numeric range.
*   The goal is to create **balanced classes** — each class should have a roughly similar number of records.
*   Poorly defined bins may lead to imbalanced data, which will require extra work later to correct.


✅ Task:
Add a new column to your DataFrame named `Classification_Label` containing your generated class labels.

💡 Hint: You can use your own if clauses or use functions like `pd.cut()` or `pd.qcut()` to bin continuous data efficiently.

````python
### BEGIN SOLUTION
bins, labels = pd.qcut(data['Regression_Label'], q=5, retbins=True)
print("Bin boundaries for each label:")
for i in range(len(labels) - 1):
    print(f"Label {i}: [{labels[i]:.2f}, {labels[i + 1]:.2f})")
data['Classification_Label'] = bins
### END SOLUTION
````

````python
#If you wrote the code above correctly, running the following block should generate a histogram of your discrete label.
#Do not change the code in this block.

data["Classification_Label"].value_counts().plot(kind='bar')

plt.xlabel("Classification Label")
plt.ylabel("Count")
plt.title("Count of Unique Values in Classification_Label")
plt.xticks(rotation=45)
plt.show()
````
